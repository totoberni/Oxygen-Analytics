#### <span style="color:green"> APPLE - MLR </span>

```python
y = mci['aapl_close']
X = mci[['mfctg_index', 'CCI_index']]
```

```python
# Specify the model.
multi = LinearRegression()  

# Fit the model.
multi.fit(X, y)
```

```python
# Checking the value of R-squared, intercept and coefficients.
print("R-squared: ", multi.score(X, y))
print("Intercept: ", multi.intercept_)
print("Coefficients:")

list(zip(X, multi.coef_))
```

```python
# Make predictions.
New_Value1 = 102
New_Value2 = 78

print ('Predicted Value: \n', multi.predict([[New_Value1 ,New_Value2]]))  
```

```python
# Split the data into training = 0.7 and testing = 0.3 subsets.
x_train, x_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.3,
                                                    random_state=42)
```

```python
# Training the model using the 'statsmodel' OLS library.
# Fit the model with the added constant.
model = sm.OLS(y_train, sm.add_constant(x_train)).fit()

# Set the predicted response vector.
Y_pred = model.predict(sm.add_constant(x_test)) 

# Call a summary of the model.
print_model = model.summary()

# Print the summary.
print(print_model)  
```

#### <span style="color:green"> Above OLS table generated very high condition number because macro retail indicator was used.  To reduce the probability that high correlation existed, this metric was removed and VIF conducted which at a value of 1  means no multi-collinearity </span>

```python
print(multi.score(x_train, y_train)*100)
```

```python
# Add a constant.
x_temp = sm.add_constant(X)  

# Create an empty DataFrame. 
vif = pd.DataFrame() 

# Calculate the 'vif' for each value.
vif["VIF Factor"] = [variance_inflation_factor(x_temp.values, 
                                               i) for i in range(x_temp.values.shape[1])]  


# Create the feature columns.
vif['features'] = x_temp.columns  

# Print the values to two decimal points.
print(vif.round(2))
```

#### <span style="color:green"> Durbin-Watson test of 2.071 indicates no auto-correlation (close to 2.0) and VIF at 1.0 which suggests no multicollinearity</span>
